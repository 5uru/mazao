```python
import csv
import base64
from PIL import Image
import io

def image_to_base64(image_path):
    with Image.open(image_path) as img:
        # Convert image to RGB mode if it's not
        if img.mode != 'RGB':
            img = img.convert('RGB')

        # Resize image if it's too large (optional)
        # img.thumbnail((800, 800))

        buffered = io.BytesIO()
        img.save(buffered, format="JPEG")
        return base64.b64encode(buffered.getvalue()).decode('utf-8')
```

```python

for file_name in file_list:
    try:
        image_directory = f"/Users/jonathansuru/PycharmProjects/mazao/datasets/{file_name}"
        output_csv = f"{file_name}.csv"
        file_name_split = file_name.split('_')
        with open(output_csv, 'w', newline='') as csvfile:
            csv_writer = csv.writer(csvfile)
            csv_writer.writerow(['plant_name', 'disease_name', 'image_data'])  # CSV header
        
            for filename in os.listdir(image_directory):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                    file_path = os.path.join(image_directory, filename)
                    try:
                        base64_image = image_to_base64(file_path)
                        csv_writer.writerow([file_name_split[0], " ".join(file_name_split[1:]), base64_image])
                        print(f"Processed: {filename}")
                    except Exception as e:
                        print(f"Error processing {filename}: {str(e)}")
        
        print(f"CSV file '{output_csv}' has been created with image data.")
    except Exception as e:
        print(f"Error processing {file_name}: {str(e)}")
```

```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
import glob

def split_csv_file(file_path, train_size=0.6, val_size=0.2, test_size=0.2):
    # Read the CSV file
    df = pd.read_csv(file_path)

    # Ensure the CSV has the expected columns
    expected_columns = ['plant_name', 'disease_name', 'image_data']
    if not all(col in df.columns for col in expected_columns):
        raise ValueError(f"CSV file {file_path} is missing one or more expected columns: {expected_columns}")

    # Perform the split
    train_val, test = train_test_split(df, test_size=test_size, stratify=df['disease_name'], random_state=42)
    train, val = train_test_split(train_val, test_size=val_size/(train_size+val_size), stratify=train_val['disease_name'], random_state=42)

    # Add the 'split' column
    train['split'] = 'train'
    val['split'] = 'validation'
    test['split'] = 'test'

    # Combine the datasets
    result = pd.concat([train, val, test])

    # Shuffle the result
    result = result.sample(frac=1, random_state=42).reset_index(drop=True)

    return result
```

```python
def process_csv_files(directory):
    # Get all CSV files in the directory
    csv_files = glob.glob(os.path.join(directory, '*.csv'))

    for file_path in csv_files:
        try:
            # Split the file
            result_df = split_csv_file(file_path)

            # Generate the output file name
            file_name = os.path.basename(file_path)
            output_file = os.path.join(directory, f'split_{file_name}')

            # Save the result
            result_df.to_csv(output_file, index=False)
            print(f"Processed {file_name}. Output saved to {output_file}")
        except Exception as e:
            print(f"Error processing {file_name}: {str(e)}")
```

```python
def concatenate_csv_files(directory, output_file):
    # Get all CSV files in the directory
    csv_files = glob.glob(os.path.join(directory, '*.csv'))

    # Sort the files to ensure a consistent order
    csv_files.sort()

    # List to hold all dataframes
    dataframes = []

    # Read each CSV file and append to the list
    for file in csv_files:
        try:
            df = pd.read_csv(file)
            # Ensure the CSV has the expected columns
            expected_columns = ['plant_name', 'disease_name', 'image_data', 'split']
            if not all(col in df.columns for col in expected_columns):
                print(f"Warning: {file} is missing one or more expected columns. Skipping.")
                continue
            dataframes.append(df)
            print(f"Processed: {file}")
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")

    # Concatenate all dataframes
    if dataframes:
        result = pd.concat(dataframes, ignore_index=True)

        # Save the concatenated result
        result.to_csv(output_file, index=False)
        print(f"Concatenation complete. Output saved to {output_file}")
        print(f"Total rows in concatenated file: {len(result)}")
    else:
        print("No valid CSV files found to concatenate.")
```

```python
# Directory containing your CSV files
csv_directory = "/Users/jonathansuru/PycharmProjects/mazao/dataset2"
# Output file path
output_file = "/datasets/data.csv"
```

```python
# Concatenate all CSV files in the directory
concatenate_csv_files(csv_directory, output_file)
```

```python
def delete_csv_files(directory):
    # Get all CSV files in the directory
    csv_files = glob.glob(os.path.join(directory, '*.csv'))

    if not csv_files:
        print(f"No CSV files found in {directory}")
        return

    # Delete the files
    deleted_count = 0
    for file in csv_files:
        try:
            os.remove(file)
            print(f"Deleted: {os.path.basename(file)}")
            deleted_count += 1
        except Exception as e:
            print(f"Error deleting {os.path.basename(file)}: {str(e)}")

    print(f"\nOperation complete. {deleted_count} CSV files were deleted.")
```
